#### 1. Rappels sur les structures de données déjà étudiées en Première <a class="anchor" id="section_1_1"></a>

##### a. Les Tableaux <a class="anchor" id="section_1_1_1"></a>

<em>définition :</em> Un <strong>tableau</strong> est une structure de données représentant une séquence finie d'éléments auxquels on peut accéder par leur position dans la séquence, appelée indice.

```python
tableau=[7,12,24,8,15]
```

Ceci est un tableau unidimensionnel. Il existe aussi des tableaux multidimensionnels dont chaque élément est lui-même un tableau. A noter que l'indexation d'un tableau de n éléments commence à 0 (premier élément) et se termine à n-1 (dernier élément).

```python
tableau[5]
```

Un tableau se caractérise par :
<ul>
    <li> le type des éléments qu'il contient (entiers, carctères,tableaux...)
    <li> le nombre d'éléments qu'il contient (sa taille)
</ul>


##### b. Les dictionnaires <a class="anchor" id="section_1_1_2"></a>

<em> définition : </em> Un <strong>dictionnaire</strong> est une structure de données dans laquelle des clés sont associés à des valeurs. Il garde en mémoire des informations du type (clé,valeur). Le but d'un dictionnaire est d'accéder rapidement à une valeur à partir de sa clé.

```python
capitale={"la France" : "Paris","la Belgique" : "Bruxelles","l'Allemagne" : "Berlin" }

print(capitale["la France"])

for cle in capitale :
    print("La capitale de {} est {}.".format(cle,capitale[cle]))
```

##### c. Les listes <a class="anchor" id="section_1_1_3"></a>

<em> définition : </em> Une <strong>liste</strong> est une structure de données regroupant des éléments ordonnés qu'il est possible de manipuler individuellement (accès, ajout, suppression et modification).

```python
maliste=[]
maliste.append("J'adore")
maliste.append("la NSI")
print(maliste)
```

<em>Remarques :</em>
<ul>
    <li> La taille d'un tableau est statique, contrairement à celle d'une liste.
    <li> Le langage Python permet de confondre les deux notions. Quand on programme en Python, on parlera de tableaux ou de listes indifféremment.
</ul>


#### 2. De nouvelles structures : les piles et les files <a class="anchor" id="section_1_2"></a>

##### a. Des donnéees en ligne : les sacs <a class="anchor" id="section_1_2_1"></a>

Les piles et les files sont des exemples de structures de données que nous appelerons des sacs. Un sac offre quatre opérations élémentaires :
<ol>
    <li> créer un sac vide ;
    <li> tester si le sac est vide ;
    <li> ajouter un élément dans le sac ;
    <li> retirer un élément du sac (attention tenter de retirer un élément d'un sac vide provoque une erreur).
</ol>
Un sac se modifie au cours du temps, il a donc une structure impérative.
<br>

##### b. Piles et files : des sacs particuliers <a class="anchor" id="section_1_2_2"></a>

Piles et Files se distinguent par la relation entre éléments ajoutés et éléments retirés.
<ul>
    <li> Dans le cas des <strong>piles</strong>, c'est le <strong>dernier</strong> élément ajouté qui est retiré.
    <li> Dans le cas des <strong>files</strong>, c'est le <strong>premier</strong> élément ajouté qui est retiré.
    </ul>
On dit que :
        <ul>
    <li> la <strong>pile</strong> est une structure <strong>LIFO</strong> (last-in first-out),
    <li> la <strong>file</strong> est une structure <strong>FIFO</strong> (fisrt-in first-out).
    </ul>
<em>Remarque :</em> Pour bien se représenter la situation : 
<ul>
    <li>on associe la structure pile à l'image d'unepile d'assiettes, dans laquelle chaque nouvelle assiette est ajouté au-dessus des précédentes, et où l'assiette retirée est systématiquement celle du sommet.
<li>on associe la structure pile à l'image d'une file d'attente, dans laquelle des personnes arrivent à tour de rôle, patientent , et sont servies dans leur ordre d'arrivée.

#### 4. La structure Pile en Python (première version)<a class="anchor" id="section_1_4"></a>

Quand on ajoute un élément à une pile, on dit que l'on empile ; quand on retire un élément d'une liste, on dit que l'on dépile.
<br>
En Python, on peut implémenter une pile à l'aide d'une liste.
<ul>
    <li> <code>p.append(v)</code> ajoute <code>v</code> au sommet de la pile (à la fin de la liste) :
    <li> <code>p[-1]</code> est l'élément au sommet de la pile (c'est le dernier élément de la liste) ;
    <li> <code>p.pop</code> retire l'élément au sommet de la pile (le dernier élément de la liste). A noter que <code>p.pop(n)</code> supprime l'élément d'indice n.

#### 4. La structure File en Python (première version)<a class="anchor" id="section_1_5"></a>

Les listes sont mal adaptées pour représenter en Python des files car l'ajout d'éléments se fait toujours en fin de liste, et non au début. C'est la raison pour laquelle on pourra utiliser la fonction <code>deque</code> du module <code>collections</code>. La syntaxe de cette fonction est :
<code> file = deque([ liste de nombres séparés par des virgules ])</code> ou :
</code>    file = deque(( liste de nombres séparés par des virgules ))</code>
Ainsi,
<center> <code>file=deque( (3,2,1) )</code></center>
représente la file dont le premier élément (tête de liste) est "1" et dont le dernier élément enfilé est "3".
On utilisera alors :
<ul>
    <li> <code>file.appendleft(élément)</code>pour enfiler un élément ;
    <li> <code>file.pop()</code> pour défiler la tête de file.

```python
from collections import deque
f=deque((3,2,1)) # initialise la file
f.append(4) # ajoute la valeur 4
print(f) # affiche la file
print(f.pop()) # affiche la tête de file et défile
print(f)
```

<h4>Les classes</h4>    
   <p align="justify"> Une classe regroupe des fonctions et des attributs qui définissent un objet. On appelle par ailleurs les fonctions d'une classe des " méthodes ". Une classe est un modèle qui permet de créer des objets de ce type.
<ul>
   <li>Les objets de cette classe sont appelés instances.
    <li>Les variables propres à chaque éléments de cette classe sont appelés attributs.
    <li>Les fonctions propres à chaque élément de cette classe sont appelées méthodes.
</ul>
L’ensemble des méthodes accessibles à un utilisateur définit l’interface de la classe.<br>
   Ex : créons une classe <em>Voiture</em>. Le code sera le suivant :</p>

```python
class Voiture:
    def __init__(self):
        self.nom = "Ferrari"
```

La variable self est toujours le premier paramètre d’une méthode. Il désigne l’objet lui même.

<h4>Les objets</h4>    
   <p align="justify"> Un objet est une instance d'une classe . On peut créer autant d'objets que l'on désire avec une classe .
<br>Créons maintenant notre voiture :</p>

```python
>>> ma_voiture = Voiture()
```

<h4>Les attributs de classe</h4>    
   <p align="justify"> Les attributs de classe permettent de stocker des informations au niveau de la classe. Elle sont similaires aux variables.<br>
Dans notre exemple :</p>

```python
>>> ma_voiture = Voiture()
>>> ma_voiture.nom
'Ferrari'
```

<h4>Les méthodes</h4>    
   <p align="justify"> Les méthodes sont des fonctions définies dans une classe.<br>
Créons une nouvelle méthode dans notre classe voiture :</p>

```python
class Voiture:
    def __init__(self):
        self.nom = "Ferrari"
    def donne_moi_le_modele(self):
        return "250"
```

Commentaires sur le code de la classe.
<ul>
<li>    Ligne 1. On utilise une majuscule pour un nom de classe.
<li>    Lignes 2 à 5. La définition de la classe se termine quand on revient à l’indentation de départ.
<li>    Ligne 2, ligne 4. définition d’une méthode
<li>    La variable self est toujours le premier paramètre d’une méthode. Il désigne l’objet lui même.
<li>    Ligne 3 : un attribut.
<li>    Ligne 3 : dans la classe on utilise self.qqchose pour accéder à un attribut ou une méthode.
<li>    La méthode __init__ est appelée “constructeur”, c’est elle qui est exécutée quand vous exécutez l’instruction Voiture(...). Elle crée un objet du type Voiture
</ul>

   <p align="justify"> Utilisons la méthode <code>donne_moi_le_modèle</code> :</p>

```python
>>> ma_voiture=Voiture()
>>> ma_voiture.donne_moi_le_modele()
'250'
```
